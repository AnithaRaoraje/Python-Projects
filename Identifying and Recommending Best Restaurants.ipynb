### 5. Find out the ratio between restaurants that allow table booking vs. those that do not allow table booking

```python
dummy = ['Has Table booking','Has Online delivery']
df = pd.get_dummies(df,columns=dummy,drop_first=True)
df.head()
```

```python
#Ration between restaurants allowing table booking and those which dont
table_booking = df[df['Has Table booking_Yes']==1]['Restaurant ID'].count()
table_booking
```

```python
notable_booking =df[df['Has Table booking_Yes']==0]['Restaurant ID'].count()
notable_booking
```

```python
print('Ratio between restaurants that allow table booking vs. those that do not allow table booking: ',
      round((table_booking/notable_booking),2))
```

### 6. Find out the percentage of restaurants providing online delivery

```python
online_del = df[df['Has Online delivery_Yes'] == 1]['Restaurant ID'].count()
online_del
```

```python
no_online_del = df[df['Has Online delivery_Yes'] == 0]['Restaurant ID'].count()
no_online_del
```

```python
print('Percentage of restaurants providing online delivery : {} %'.format((round(online_del/len(df),3)*100)))
```
